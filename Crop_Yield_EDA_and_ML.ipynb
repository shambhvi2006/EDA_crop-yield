```python
def preprocess_dataset(df):
    df_clean = df.copy()
    sources = {
        'rain': 'Rain',
        'tawi': 'Tawi',
        'well': 'Well'
    }

    for keyword, col in sources.items():
        df_clean[col] = df_clean['Water'].astype(str).str.lower().str.contains(keyword)

    if 'Water' in df_clean.columns:
        df_clean = df_clean.drop(columns=['Water'])
    if 'Soil_Addition' in df_clean.columns:
        df_clean['Soil_Addition'] = (
            df_clean['Soil_Addition']
            .astype(str)
            .str.lower()
            .str.strip()
            .replace({
                'compost and manures': 'compost and manure',
                'compost & manure': 'compost and manure',
                'compast and manures': 'compost and manure',
                'compact and manures': 'compost and manure'

            })
        )
        soil_encoded = pd.get_dummies(df_clean['Soil_Addition'], prefix="Soil_Addition")
        df_clean = pd.concat([df_clean.drop(columns=['Soil_Addition']), soil_encoded], axis=1)

    yes_no_cols = [
        col for col in df_clean.columns
        if col.lower() != "name" and
        df_clean[col].astype(str).str.contains("yes|no", case=False, na=False).any()
    ]

    for col in yes_no_cols:
        df_clean[col] = (
            df_clean[col]
            .astype(str)
            .str.lower()
            .str.strip()
            .apply(lambda x: "yes" if "yes" in x else ("no" if "no" in x else None))
        )
        df_clean[col] = df_clean[col].map({"yes": True, "no": False})

    return df_clean
```

```python
import numpy as np
import pandas as pd

new=df_cleann.copy()
# Function to clean numeric values by stripping non-numeric chars and converting to float
def clean_numeric(col):
    cleaned_col = (
        col.astype(str)
        .str.replace(r"[^0-9.\-]", "", regex=True)  # keep only numbers, dot, minus
        .str.replace(r"(\.)+", r".", regex=True) # Replace multiple dots with a single dot
        .str.replace(r"^\.", "", regex=True) # Remove leading dot
        .str.replace(r"\.$", "", regex=True) # remove trailing dot
        .replace("", np.nan)  # empty strings -> NaN
    )
    # Handle cases like '2.3.46' by keeping only the first decimal point
    cleaned_col = cleaned_col.apply(lambda x: x.split('.', 1)[0] + '.' + ''.join(x.split('.')[1:]) if isinstance(x, str) and x.count('.') > 1 else x)
    # print(f"Values before converting to float for column {col.name}:\n{cleaned_col.unique()}") # Add this line to inspect values
    return cleaned_col.astype(float) # Convert to float, coercing errors to NaN

# Apply cleaning to all potentially problematic numeric columns
numeric_cols_to_clean = [
    "Soil pH",
    "Electrical Conductivity",
    "Organic carbon (OC)",
    "Nitrogen",
    "Phosphorous",
    "Potassium",
    "Sulphur",
    "Boron",
    "Zinc",
    "Iron",
    "Humidity",
    "Annual Rainfall",
    "Data of production",
    "Fertilizers", # Added for imputation
    "Metals and pollutants" # Added for imputation
]

for col in numeric_cols_to_clean:
    if col in new.columns:
        new[col] = clean_numeric(new[col])

# Impute remaining NaN values in numeric columns with the mean
for col in new.select_dtypes(include=['float64', 'int64']).columns: # Impute all numeric columns
    if new[col].isnull().any():
        new[col] = new[col].fillna(new[col].mean())

print(new.head())
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, HistGradientBoostingRegressor, ExtraTreesRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

models = {
    "Linear Regression": LinearRegression(),
    "Ridge": Ridge(),
    "Lasso": Lasso(),
    "ElasticNet": ElasticNet(),
    "Random Forest Regressor": RandomForestRegressor(),
    "Gradient Boosting Regressor": GradientBoostingRegressor(),
    "Hist Gradient Boosting Regressor": HistGradientBoostingRegressor(),
    "Extra Trees Regressor": ExtraTreesRegressor(),
    "SVM Regressor": SVR(),
    "XGBoost Regressor": XGBRegressor(use_label_encoder=False, eval_metric="rmse"),
    "LightGBM Regressor": LGBMRegressor(),
    "CatBoost Regressor": CatBoostRegressor(verbose=0)
}

results = []

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    results.append([name, mse, mae, r2])
import pandas as pd
results_df = pd.DataFrame(results, columns=["Model", "MSE", "MAE", "R2 Score"])
print(results_df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.bar(results_df["Model"], results_df["R2 Score"])
plt.xticks(rotation=45, ha="right")
plt.ylabel("RÂ² Score")
plt.title("Model Comparison (Higher is Better)")
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout

# Build MLP model
dl_model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dense(1)   # single neuron (no activation) for regression
])

# Compile model
dl_model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Train model
history = dl_model.fit(
    X_train, y_train,
    epochs=50, batch_size=32,
    validation_data=(X_test, y_test),
    verbose=1
)

# Evaluate on test set
mse, mae = dl_model.evaluate(X_test, y_test)
print(f"Deep Learning Regressor -> MSE: {mse:.4f}, MAE:Â {mae:.4f}")
```

```python
# Calculate R2 Score for DL
from sklearn.metrics import r2_score
y_pred_dl = dl_model.predict(X_test)
r2_dl = r2_score(y_test, y_pred_dl)

# Add DL results to results_df
results_df.loc[len(results_df)] = ["Deep Learning (MLP)", mse, mae, r2_dl]

print(results_df)
```

```python
# Create SHAP explainer
explainer = shap.Explainer(model, X_train)
shap_values = explainer(X_test)

# Global feature importance
shap.summary_plot(shap_values, X_test)

# Dependence plot (for one feature)
import numpy as np

# Get mean absolute SHAP value for each feature
feature_importance = np.abs(shap_values.values).mean(axis=0)

# Sort features by importance
sorted_idx = np.argsort(feature_importance)[::-1]  # descending order

# Pick top N features
N = 5  # change this number if you want more/less features
top_features = X_train.columns[sorted_idx[:N]].tolist()

print("Top", N, "features:", top_features)

# Plot dependence plots for top N features
for f in top_features:
    print(f"\nDependence plot for {f}")
    shap.dependence_plot(f, shap_values.values, X_test)

```
